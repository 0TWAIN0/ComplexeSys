```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

N=5000
L=1000
lya = np.zeros(L)
x = np.zeros(L)+0.1
r = np.linspace(2.01,4,L)

for i in range(0,N):
    lya = lya + np.log(abs(r*(1-2*x)))
    x = r*x*(1-x)
    
lya = lya / N

plt.figure()
plt.plot(r,lya)
plt.ylim([-2,1])
plt.show()
    
```
