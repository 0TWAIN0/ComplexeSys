```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

N=5000
L=1000
lya = np.zeros(L)
x = np.zeros(L)+0.1
r = np.linspace(2.01,4,L)

for i in range(0,N):
    lya = lya + np.log(abs(r*(1-2*x)))
    x = r*x*(1-x)
    
lya = lya / N

plt.figure()
plt.plot(r,lya)
plt.ylim([-2,1])
plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

N = 100000
b = np.array([2,2.5,3,np.pi])
x = np.array([0.1,0.1,0.1,0.1])
L = 100
counting = np.zeros((L,4))

for i in range(0,N):
    counting[np.floor(x[0] * L),0] += 1
    counting[np.floor(x[1] * L),1] += 1
    counting[np.floor(x[2] * L),2] += 1
    counting[np.floor(x[3] * L),3] += 1
    x = np.mod(b*x + (np.random.rand()*2-1)/1000 ,1)

plt.figure()
plt.plot(np.linspace(0,1,L),counting[:,0]/N *L,)
plt.ylim([0,1.5])
plt.show()
plt.figure()
plt.plot(np.linspace(0,1,L),counting[:,1]/N *L,)
plt.ylim([0,1.5])
plt.show()
plt.figure()
plt.plot(np.linspace(0,1,L),counting[:,2]/N *L,)
plt.ylim([0,1.5])
plt.show()
plt.figure()
plt.plot(np.linspace(0,1,L),counting[:,3]/N *L,)
plt.ylim([0,1.5])
plt.show()
```
