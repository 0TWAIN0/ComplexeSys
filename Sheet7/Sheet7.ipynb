```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

Nmax = 100
L = 50
K = 0.1

def std_map(xp, K):
    return np.array([np.mod(xp[:,0]+xp[:,1],1), 
                     np.mod(xp[:,1]+K*np.sin(2*np.pi*(xp[:,0]+xp[:,1]))/(2*np.pi),1) -0.5]).transpose()

x,p = np.meshgrid(np.linspace(0,1,L),np.linspace(-0.5,0.5,L))
grid = np.zeros((L,L,2))
grid[:,:,0] = x
grid[:,:,1] = p

xp = grid.reshape(L*L,2)
plt.figure()
plt.ylim([-0.5,0.5])
for n in range(1,Nmax+1):
    plt.plot(xp[:,0],xp[:,1],',k',alpha=0.3)
    xp = std_map(xp,K)
plt.show()
```
