```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

ITER = 100 # iterations
LGRID = 50 # grid length
K = 1

def stdmap(X,P,K):
    return np.mod(X + P,1), \
           np.mod(P + K/(2 * np.pi) * np.sin(2 * np.pi * (X + P)) + 0.5,1) -0.5

x,p = np.meshgrid(np.linspace(0,1,LGRID),np.linspace(-0.5,0.5,LGRID))

plt.figure()
plt.ylim([-0.5,0.5])
plt.xlabel('x')
plt.ylabel('p')

for n in range(ITER):
    plt.plot(x,p,',k',alpha=0.3)
    x, p = stdmap(x, p, K)
    
plt.show()
```
